```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CitiesData.csv')
```

```python
df['Cuisines'].value_counts()
```

```python
sns.set_context("paper", font_scale = 2, rc = {"font.size": 14,"axes.titlesize": 25,"axes.labelsize": 16})   
sns.catplot(data = df, kind = 'count', x = 'Cuisines',order =df['Cuisines'].value_counts(normalize=True).index[:15])
plt.title('Popular cuisines in 5 cities')
plt.xticks(rotation = 90)
plt.show()
```
