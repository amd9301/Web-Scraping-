```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Swiggy Bangalore Outlet Details.csv')
```

```python
dataset = df.copy()
df.head()
```

```python
df.info()
```

```python
df['Rating'].value_counts()
```

```python
df.isnull().sum()
```

```python
sns.set_context("paper", font_scale = 2, rc = {"font.size": 20,"axes.titlesize": 25,"axes.labelsize": 20})   
sns.catplot(data = df, kind = 'count', x = 'Rating')
plt.title('Rating chart')
plt.show()
```

```python
a=list(df['Rating'])
for i in range(0,len(a)):
    if(a[i]=='nan'):
        a[i]='unrated'
        
    elif a[i]=='-':
        a[i]='unrated'
    elif a[i]=='NEW':
        a[i]='unrated'
        
```

```python
df['Rating']=a
```

```python
df['Rating'].value_counts()
```

```python
# Plotting the number of restaurants for each rating
sns.set_context("paper", font_scale = 2, rc = {"font.size": 20,"axes.titlesize": 25,"axes.labelsize": 20})   
b = sns.catplot(data = df, kind = 'count', x = 'Rating', order = df['Rating'].value_counts().index)
plt.title('Number of restaurants for each rating')
b.set_xticklabels(rotation = 90)
plt.show()
```

```python
sns.set_context("paper", font_scale = 2, rc = {"font.size": 14,"axes.titlesize": 25,"axes.labelsize": 16})   
sns.catplot(data = df, kind = 'count', x = 'Cuisine',order =df['Cuisine'].value_counts(normalize=True).index[:10])
plt.title('Top 10 cuisine types in Bangalore')
plt.xticks(rotation = 90)
plt.show()
```
